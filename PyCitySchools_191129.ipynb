```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data = "Resources/schools_complete.csv"
student_data = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data)
student_data = pd.read_csv(student_data)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data_complete.head()
```

* Calculate the average reading score

```python
school_count = len(school_data_complete["school_name"].unique())
student_count = school_data_complete["Student ID"].count()
print('The total number of schools is '+ str(school_count))
print('The total number of students is '+ str(student_count))

total_budget_raw = school_data["budget"].sum()
total_budget = "$"+"{:,}".format(total_budget_raw)
print('The total budget is '+ str(total_budget))

avg_math_score = round(student_data["math_score"].mean(),2)
avg_reading_score = round(student_data["reading_score"].mean(),2)
print('The average math score is '+str(avg_math_score))
print('The average reading score is '+str(avg_reading_score))

overall_passing_score = (avg_math_score + avg_reading_score)/2
print('The overall passing score is '+str(overall_passing_rate))

passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = round(passing_math_count/float(student_count) * 100,2)
print('The percentage of students with a passing math score is ' + str(passing_math_percentage))

passing_reading_count = school_data_complete [(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = round(passing_reading_count/float(student_count) * 100,2)
print('The percentage of students with a passing reading score is ' + str(passing_reading_percentage))

dis_sum = pd.DataFrame({"Total Schools": [school_count],
                              "Total Students": [student_count],
                              "Total Budget": [total_budget],
                              "Average Math Score": [avg_reading_score],
                              "Average Reading Score": [avg_math_score],
                              "Overall Passing Score": [overall_passing_rate], 
                              "Math Passing Rate %": [passing_math_percentage],
                              "Reading Passing Rate %":[passing_reading_percentage]})

dis_sum
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
#groups by school
sch_group = school_data_complete.set_index('school_name').groupby(['school_name'])

#school types
sch_types = school_data.set_index('school_name')['type']

#Total Students
total_stu_sch = sch_group['Student ID'].count()

#Total School Budget
sch_budget = school_data.set_index('school_name')['budget']

#per student budget
stu_budget = school_data.set_index('school_name')['budget']/school_data.set_index('school_name')['size']

#avg scores by school
avg_math = round(sch_group['math_score'].mean(),2)
avg_reading = round(sch_group['reading_score'].mean(),2)

# % passing scores
pass_math_raw = school_data_complete[school_data_complete['math_score'] >= 70].groupby('school_name')['Student ID'].count()/total_stu_sch 
pass_math = round((pass_math_raw) * 100,2)
pass_reading_raw = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('school_name')['Student ID'].count()/total_stu_sch 
pass_reading = round((pass_reading_raw) * 100,2)
overall_rate = round((pass_math_raw + pass_reading_raw)/2 *100,2)


sch_sum = pd.DataFrame({"School Type": sch_types,"Total Students": total_stu_sch, "Total School Budget $": sch_budget,
                        "Per Student Budget $": stu_budget,"Average Math Score": avg_math,
                            "Average Reading Score": avg_reading,"Math Passing %": pass_math,
                            "Reading Passing %": pass_reading, "Overall Passing Rate %": overall_rate})

sch_sum.style.format({ "Total School Budget": "${:,}", "Per Student Budget": "${:,}",
                      "Average Math Score": "{:,}", "Average Reading Score": "{:,}"})

sch_sum
```
